```
from google.colab import drive
import pandas as pd

# Montar Google Drive
drive.mount('/content/drive')

# Ruta del archivo (ajusta según tu ubicación en Google Drive)
file_path = '/content/drive/My Drive/country_vaccinations.csv'

# Cargar el archivo CSV
df = pd.read_csv(file_path)

# Ver las primeras filas para familiarizarnos con los datos
print("Primeras filas del archivo:")
print(df.head())

# Ver las columnas del DataFrame para confirmar el nombre correcto
print("\nColumnas del DataFrame:")
print(df.columns)

# Información general sobre los datos (tipos de columnas y valores nulos)
print("\nInformación del dataframe:")
df.info()

# Estadísticas descriptivas básicas
print("\nEstadísticas descriptivas:")
print(df.describe())

# Asegurarse de que la columna de fecha esté en formato datetime
df['date'] = pd.to_datetime(df['date'])

# Análisis básico
# 1. Cantidad de vacunas aplicadas por compañía
# Ajusta 'vaccine' al nombre correcto de la columna, por ejemplo, 'vaccine_name'
vaccines_by_company = df.groupby('vaccines')['total_vaccinations'].max()
print("\nVacunas aplicadas por compañía:")
print(vaccines_by_company)

# 2. Cantidad de vacunas aplicadas en todo el mundo
total_vaccinations_world = df['total_vaccinations'].max()
print("\nVacunas aplicadas en todo el mundo:", total_vaccinations_world)

# 3. Promedio de vacunas aplicadas por país
avg_vaccinations_per_country = df.groupby('country')['total_vaccinations'].mean()
print("\nPromedio de vacunas aplicadas por país:")
print(avg_vaccinations_per_country.head())

# 4. Cantidad de vacunas aplicadas el 29/01/2021
vaccinations_on_29_jan = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()
print("\nVacunas aplicadas el 29/01/2021:", vaccinations_on_29_jan)


# 6. DataFrame con los registros antes del 20 de diciembre de 2020
before_20_dec = df[df['date'] < '2020-12-20']
print("\nRegistros antes del 20/12/2020 (primeros 5 registros):")
print(before_20_dec.head())

# 7. DataFrame con los registros de la vacuna Pfizer
pfizer_data = df[df['vaccines'] == 'Pfizer']
print("\nRegistros de la vacuna Pfizer (primeros 5 registros):")
print(pfizer_data.head())

# Guardar los resultados en archivos CSV
output_path_vaccines_by_company = '/content/vaccines_by_company.csv'
vaccines_by_company.to_csv(output_path_vaccines_by_company)

output_path_total_vaccinations = '/content/total_vaccinations_world.csv'
total_vaccinations_world_df = pd.DataFrame([total_vaccinations_world], columns=['total_vaccinations'])
total_vaccinations_world_df.to_csv(output_path_total_vaccinations, index=False)

output_path_avg_vaccinations_per_country = '/content/avg_vaccinations_per_country.csv'
avg_vaccinations_per_country.to_csv(output_path_avg_vaccinations_per_country)

output_path_vaccinations_on_29_jan = '/content/vaccinations_on_29_jan.csv'
vaccinations_on_29_jan_df = pd.DataFrame([vaccinations_on_29_jan], columns=['vaccinations_on_29_jan'])
vaccinations_on_29_jan_df.to_csv(output_path_vaccinations_on_29_jan, index=False)

output_path_before_20_dec = '/content/before_20_dec.csv'
before_20_dec.to_csv(output_path_before_20_dec, index=False)

output_path_pfizer_data = '/content/pfizer_data.csv'
pfizer_data.to_csv(output_path_pfizer_data, index=False)

output_path_with_differences = '/content/df_with_differences.csv'
df.to_csv(output_path_with_differences, index=False)

print("\nLos archivos CSV se han guardado en las siguientes rutas:")
print("Vacunas aplicadas por compañía:", output_path_vaccines_by_company)
print("Vacunas aplicadas en todo el mundo:", output_path_total_vaccinations)
print("Promedio de vacunas aplicadas por país:", output_path_avg_vaccinations_per_country)
print("Vacunas aplicadas el 29/01/2021:", output_path_vaccinations_on_29_jan)
print("Registros antes del 20/12/2020:", output_path_before_20_dec)
print("Registros de la vacuna Pfizer:", output_path_pfizer_data)
print("DataFrame con la diferencia de fechas:", output_path_with_differences)
```
